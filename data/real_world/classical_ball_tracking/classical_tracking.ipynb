```python
dataset_folder = '../classical_tracking_dataset'
image_dir = os.path.join(dataset_folder, 'frames')
image_file = '13-38-07_17_0085.png'
image_path = os.path.join(image_dir, image_file)
image = Image.open(image_path)
def transform_scale_polar(img):
    """
    choose a scale of 360 px --> to get easy 360 deg
    """
    # image resize
    ht, wd = img.shape[:2]
    scale = 360/ht
    dim = (int(scale*wd), 360)
    img_resize = cv2.resize(img, dim)
    moments = cv2.moments(img_resize)
    cx_resize = moments['m10']/moments['m00']
    cy_resize = moments['m01']/moments['m00']
    h, w = img_resize.shape[:2]
    radius = min([h, w])//2
    # M = w / log(radius)
    img_polar = cv2.linearPolar(img_resize, (cx_resize,cy_resize), radius, cv2.INTER_LINEAR)
    return img_polar
```

```python
def crop_white_margins(image):
    """
    Automatically crops white margins from an image.
    """
    # Convert to binary: assume near-white areas (>=240) are background
    _, bin_img = cv2.threshold(image, 240, 255, cv2.THRESH_BINARY)

    # Find all non-white pixels
    coords = cv2.findNonZero(255 - bin_img)  # Invert to detect content

    # Get bounding box of non-white region
    x, y, w, h = cv2.boundingRect(coords)

    # Crop the image to this bounding box
    cropped_img = image[y:y+h-1, x:x+w]
    return cropped_img
```

```python
# convert image to grayscale
image_np = np.array(image)
image_gray = cv2.cvtColor(image_np, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray, cmap='gray')
ht, wd = image_gray.shape[:2]
image_cropped = crop_white_margins(image_gray)
scale = 360/ht
dim = (int(scale*wd), 360)
img_resize = cv2.resize(image_cropped, dim)
# Enhance contrast using CLAHE
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
contrast = clahe.apply(img_resize)
plt.figure()
plt.imshow(contrast, cmap='gray')
_, bin_img = cv2.threshold(contrast, 50, 255, cv2.THRESH_BINARY)
moments = cv2.moments(bin_img)
cx_resize = moments['m10']/moments['m00']
cy_resize = moments['m01']/moments['m00']
ht_resize, wd_resize = contrast.shape[:2]
# cx_resize, cy_resize = ht_resize - cy_resize, cx_resize
plt.figure()
plt.imshow(bin_img, cmap='gray')
plt.scatter(cx_resize, cy_resize, c='r')
h, w = contrast.shape[:2]
max_radius = min([h, w])//2
# M = w / log(radius)
img_polar = cv2.linearPolar(contrast, (cx_resize,cy_resize), max_radius, cv2.WARP_FILL_OUTLIERS)
r_circle_polar = img_polar.argmax(axis=1).mean()

plt.figure()
plt.imshow(img_polar, cmap='gray')
plt.plot([r_circle_polar, r_circle_polar], [0, 360], c='r')

# Convert r_circle to pixels
r_circle = r_circle_polar * max_radius / w

# Plot the circle in the original image
plt.figure()
plt.imshow(contrast, cmap='gray')
plt.scatter(cx_resize, cy_resize, c='r')
plt.plot([cx_resize, cx_resize + r_circle], [cy_resize, cy_resize], c='r')
plt.plot([cx_resize, cx_resize], [cy_resize, cy_resize + r_circle], c='r')
plt.plot([cx_resize, cx_resize - r_circle], [cy_resize, cy_resize], c='r')
plt.plot([cx_resize, cx_resize], [cy_resize, cy_resize - r_circle], c='r')
```

```python
# Generate an image with a circle of radius r_circle
r_circle = 30
c_x, c_y = 100, 100
circle = np.zeros_like(contrast)
for i in range(contrast.shape[0]):
    for j in range(contrast.shape[1]):
        if (i - c_x)**2 + (j - c_y)**2 < r_circle**2 and (i - c_x)**2 + (j - c_y)**2 > (r_circle - 5)**2:
            circle[i, j] = 255
plt.figure()
plt.imshow(circle, cmap='gray')
```
